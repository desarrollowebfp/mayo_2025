## Diccionarios


```python
hulk = {
    "nombre": "Hulk",
    "identidad": "Bruce Banner",
    "edad": 48,
}

print(hulk)
```

```python
ironman = dict(nombre = "Iron Man", identidad = "Tony Stark", edad = 50)
print(ironman)
```

```python
# INDEXACIÓN
ironman = { 
    'nombre': 'Iron Man',
    'identidad': 'Tony Stark',
    'edad': 50
    }

print(ironman["nombre"])

# print(ironman["ciudad"]) Esto arroja error porque no existe la clave

# Existe un método alternativo para acceder a claves de un diccionario sin que arroje error, al cual le podemos asignar un valor por defecto como segundo argumento
print(ironman.get("ciudad", "No se especifica"))
```

```python
# MODIFICAR
ironman["identidad"] = "Antonio Stark"
print(ironman)

# AÑADIR UNA PROPIEDAD
ironman["ciudad"] = "New York"
print(ironman)
```

```python
# FUNCIONES O MÉTODOS

# keys -> devuelve una vista con todas las claves
print(ironman.keys())

lista_keys = list(ironman.keys())
print(lista_keys[0])

# values -> devuelve una vista con todos los valores
print(ironman.values())

# items -> devuelve una vista con tuplas de claves y valores
print(ironman.items())
for clave, valor in ironman.items():
    print(f"{clave}: {valor}")

# update -> recibe por argumento otro diccionario y actualiza el diccionario actual
ironman.update({
    "identidad": "Tony Stark",
    "edad": 51,
    "coche": "Seat Leon"
})

print(ironman)

# setdefault -> devuelve el valor de una clave si existe, y si no existe la crea con un valor por defecto indicado
print(ironman.setdefault("pais", "Estados Unidos"))
print(ironman)

# pop -> Elimina un elemento de un diccionario y devuelve su valor, si no se indica un valor por defecto arroja una Excepción
print(ironman.pop("familia", None))
print(ironman)

# popitem -> Elimina y devuelve el ultimo par de clave y valor insertado
print(ironman.popitem())
print(ironman)

# copy -> Copia sin referencia o superficial del diccionario
ironman_copy = ironman.copy()
ironman_copy["nombre"] = "Aquaman"
print(ironman)
print(ironman_copy)

# clear -> Limpiar todas las claves y valores de un objeto
ironman_copy.clear()
print(ironman_copy)
print(ironman)
```
