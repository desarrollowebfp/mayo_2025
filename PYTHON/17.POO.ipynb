```python
class Movie:
    """
    Representa una película con información básica
    """
    def __init__(self, title, year, duration_min, genre):
        self.title = str(title)
        self.year = int(year)
        self.duration_min = int(duration_min)
        self.genre = str(genre)
        
    def show_info(self):
        return (f"La película {self.title} es del género {self.genre}");
    
    def readable_duration(self):
        hours = self.duration_min // 60
        mins = self.duration_min % 60
        return (f"{hours} h - {mins} m")
        
matrix = Movie("The Matrix", 1995, 123, "Sci-fi")
print(matrix.duration_min)
print(matrix.show_info())
print(matrix.readable_duration())

arrival = Movie("Arrival", 2016, 116, "Sci-fi")
print(arrival.show_info())
print(arrival.readable_duration())
```

```python
class Vehiculo:
    def __init__(self, nombre, color, marca):
        self.nombre = str(nombre)
        self.color = str(color)
        self.marca = str(marca)
        
    def avanzar(self):
        return f"El vehiculo {self.marca} {self.nombre} está avanzando"
    
    def detener(self):
        return f"El vehiculo {self.marca} {self.nombre} está detenido"
    
seat_ibiza = Vehiculo("Ibiza", "Gris oscuro", "Seat")
print(seat_ibiza.avanzar())

class Avion(Vehiculo):
    def __init__(self, nombre, color, marca, numero_alas):
        super().__init__(nombre, color, marca)
        self.numero_alas = int(numero_alas) 
        
    def volar(self):
        return f"El avión {self.marca} {self.nombre} está volando"
    
class Coche(Vehiculo):
    def __init__(self, nombre, color, marca):
        super().__init__(nombre, color, marca)
        
    def derrapar(self):
        return f"El coche {self.marca} {self.nombre} está derrapando"
    
boeing_737 = Avion("737", "Blanco", "Boeing", 2)
print(boeing_737.volar())
seat_ibiza = Coche("Ibiza", "Gris", "Seat")
print(seat_ibiza.derrapar())
```
