## List Comprehension


```python
# nueva = [expresion for element in iterable]
```

```python
numeros = []
for numero in range(5):
    numeros.append(numero)   
print(numeros)    

##########################

numeros_lc = [numero for numero in range(5)]
print(numeros_lc)
```

```python
cuadrados = []

for numero in range(11):
    cuadrados.append(numero*numero)
print(cuadrados)

##########################

cuadrados_lc = [numero*numero for numero in range(11)]
print(cuadrados_lc)
```

```python
pares = []
for numero in range(11):
    if numero % 2 == 0:
        pares.append(numero)
print(pares)

###################################

pares_lc = [numero for numero in range(11) if numero % 2 == 0]
print(pares_lc)
```

```python
precios = [10.0, 4.65, 100, 2.99]
precios_con_iva = [round(precio*1.21,2)for precio in precios]
print(precios_con_iva)
```

```python
nombres = ["ana", "pedro", "lucía"]
mayus = [s.upper() for s in nombres]
print(mayus)  # ['ANA', 'PEDRO', 'LUCÍA']
```
