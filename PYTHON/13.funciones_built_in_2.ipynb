### range


```python
# range(start, stop, step)
for n in range(5):
    print(n) 
    
print("---------------------")    
    
for n in range(1, 6):
    print(n)    
    
print("---------------------")    

for n in range(0, 10, 2):
    print(n)
    
print("---------------------")    

for n in range(5, 0, -1):
    print(n)
    
print("---------------------")     

list_numeros = []

for n in range(1, 21):
    list_numeros.append(n)
  
print(list_numeros)
```

### enumerate


```python
frutas = ["melocotón", "tomate", "pera", "naranja", "sandía", "melón", "kibi"]

for indice, fruta in enumerate(frutas):
    print(f"{fruta} - {indice}")
    
print("---------------------")  

for indice, fruta in enumerate(frutas, start = 1):
    print(f"{fruta} - {indice}")
```

### zip


```python
alumnos = ["Daniel", "Ezequiel", "Fermin", "Franchezca", "Helena", "Óscar", "Paco", "Rafa", "Ro", "Sergio", "Antonio"]

notas = [5,7,3,8,7,3,9,10,6,7]

for alumno, nota in zip(alumnos, notas):
    print(f"{alumno} : {nota}")
    
boletin = dict(zip(alumnos, notas))

print(boletin)
```
