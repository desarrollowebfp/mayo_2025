## Operadores


```python
# Operaciones de concatenación
saludo = "Hola, "
nombre = "Sergio"
edad = 27

mensaje = saludo + nombre + " y tengo " + str(edad) + " años"
print(mensaje)
```

```python
# Operadores de comparación
y = 10
x = 20

print(y == x)
print(y != x)
print(y > x)
print(y >= x)
print(y < 10)
print(y <= 10)

print(10 == "10")
print(10 == 10)
```

```python
# Operadores lógicos

tiene_coche = True
tiene_carnet = False

# Tienes que tener coche Y carnet
print(tiene_coche and tiene_carnet)
# Caso ficticio: si tienes coche, el carnet es opcional
print(tiene_coche or tiene_carnet)
```

```python
# Operaciones aritméticas directas
a = 10
b = 4

resultado = a + b
# Forma larga -> resultado = resultado + 5
resultado += 5
print(resultado)
```

```python
# Conversiones de tipo de dato

# Convertir de string a numero entero
numero = int("5")

# Convertir de string a numero decimal
numero_decimal = float("5.6")

# Convertir de numero a string
numero_string = str(106)

# Convertir de numero a boolean
numero_boolean = bool(654)
print(numero_boolean)
```
