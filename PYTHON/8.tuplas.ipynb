## Tuplas


```python
superheroe = ("Spider-man", "Peter Parker", "Marvel", 1962)
print(type(superheroe))
```

```python
# Las tuplas son inmutables
tupla_ejemplo = (1,2,3,4,5,6)
# tupla_ejemplo[2] = 10
print(tupla_ejemplo)
```

```python
# Formas de crear una tupla -> Si queremos crear una tupla con un solo elemento tenemos que ponerle una trailing comma al final
otro_ejemplo = (5,)
print(type(otro_ejemplo))
```

```python
# Indexación con tuplas
videojuego = ("The Legend of Zelda", "Nintendo", 1996)
print(videojuego[0])
print(videojuego[-1])
```

```python
# Saltarnos la inmutabilidad
videojuego = ("The Legend of Zelda", "Nintendo", 1996)
print(type(videojuego))

videojuego_lista = list(videojuego)
videojuego_lista[0] = "The Legend of Paco"

videojuego = tuple(videojuego_lista)
print(videojuego)
```

```python
# index -> Devuelve la posición del primer elemento del valor buscado
videojuego = ("The Legend of Zelda", "Nintendo", 1996)

print(videojuego.index("Nintendo"))

# count -> Contar cuantas veces aparece un elemento en una tupla
print(videojuego.count("Nintendo"))

# len -> Contamos el numero de elementos que hay dentro de una tupla
print(len(videojuego))

# min y max
numeros = (1,5,6,2,8,10)
print(min(numeros))
print(max(numeros))
```
