## Variables y tipos de datos


```python
# Esto es un comentario de Python
tecnologia = "Python"       # Esto es una variable con un texto

print("La variable tecnologia tiene como valor:", tecnologia)
print(f"La variable tecnologia tiene como valor: {tecnologia}")
```

```python
# Tipos de datos

# Numbers - Numeros
numero_entero = 50 # int - integer
print(type(numero_entero))
numero_decimal = 50.5 # float
print(type(numero_decimal))
numero_complejo = 2 + 3j # complex
print(type(numero_complejo))
```

```python
# Strings
nombre = "Antonio" # str - string
perro = 'Byron'
multilinea = '''
Hola, soy un
texto
multilinea
'''
ejemplo_diferentes_comillas = 'Pepito dijo: "Python es super facil"'
print(ejemplo_diferentes_comillas)
print(multilinea)
print(type(multilinea))
```

```python
# Booleans
mayor_edad = True # bool
carnet_conducir = False
print(type(mayor_edad))
```

```python
tipo_dato_none = None # NoneType
print(type(tipo_dato_none))
```
