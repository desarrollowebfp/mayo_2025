## Sets


```python
clasicos = {"Casablanca", "Titanic", "Forrest Gump", "El Padrino", "Gladiator"}
scifi = {"Matrix", "Interstellar", "Inception", "Titanic", "Avatar"}
animacion = {"Up", "Coco", "Toy Story", "Shrek", "Avatar"}
```

```python
# len
print(len(clasicos))

# copy -> Hace una copia superficial
clasicos_copiados = clasicos.copy()

# add -> añade un nuevo elemento al set
scifi.add("2001: Odisea en el espacio")
scifi.add("Matrix")
print(scifi)

# update -> mediante una lista añade varios elementos al set
animacion.update(["El castillo ambulante", "Titan A.E."])
print(animacion)

# pop -> elimina un elemento aleatorio y no se puede elegir cual
# print(clasicos.pop())

# remove -> elimina un elemento especifico y si no se encuentra da error
# clasicos.remove("Bracula II")
print(clasicos)

# discard -> elimina un elemento especifico y si no se encuentra no da error
clasicos.discard("Bracula II")

# clear -> Vacia el set por completo
clasicos.clear()
print(clasicos)
```

```python
# OPERACIONES ENTRE SETS O CONJUNTOS

clasicos = {"Casablanca", "Titanic", "Forrest Gump", "El Padrino", "Gladiator"}
scifi = {"Matrix", "Interstellar", "Inception", "Titanic", "Avatar"}
animacion = {"Up", "Coco", "Toy Story", "Shrek", "Avatar"}

# union -> |
peliculas_union = clasicos | scifi
print(peliculas_union)

# intersection -> &
peliculas_comunes = clasicos & scifi
print(peliculas_comunes)

# difference -> -
solo_clasicos = clasicos.difference(scifi)
print(solo_clasicos)

# symmetric difference -> ^
peliculas_unicas = clasicos.symmetric_difference(scifi)
print(peliculas_unicas)

# isdisjoint -> Comprueba si no tienen elementos en comun los conjuntos
print(clasicos.isdisjoint(scifi))

# issubtset y issuperset -> Comprueba si un conjunto esta dentro de otro
```
