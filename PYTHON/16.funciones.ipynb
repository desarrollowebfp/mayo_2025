### Funciones


```python
def saludar():
    print("Hola!")
    
saludar()
saludar()
```

```python
def saludar_con_nombre(nombre):
    print(f"Hola {nombre}")
    
saludar_con_nombre("Nau")
saludar_con_nombre("Ro")
saludar_con_nombre("Jose")
```

```python
def sumar_dos_numeros(numeroA, numeroB):
    total = numeroB + numeroA
    print(total)
    
sumar_dos_numeros(5, 6.80)
```

```python
def announce_year(year = 2025):
    print(f"Estamos en el año: {year}")
    
announce_year(2020)
announce_year()
```

```python
def sumar (numeroA, numeroB):
    """
    Suma dos números y retorna el resultado

    Args:
        numeroA (int): Primer número a sumar
        numeroB (int): Segundo número a sumar

    Returns:
        int: El resultado de los dos números sumados
    """
    total = numeroA + numeroB
    return total

cinco_mas_seis = sumar(5,6)
print(cinco_mas_seis)

sumar(5,7)

print(help(sumar))
```

```python
# *args -> argumentos posicionales (se guardan en una tupla)

def mostrar_nombres (*args):
   for nombre in args:
       print(nombre)
    
mostrar_nombres("Ro", "Fermin", "Rafa", "Paco", "Sergio", "Ezequiel")
```

```python
## **kwargs -> pares de clave-valor (mini-diccionario)
def mostrar_user (**kwargs):
    for clave, valor in kwargs.items():
        print(f'{clave} : {valor}')
    
mostrar_user(nombre = "Iron Man", identidad = "Tony Stark", edad = 50)
```

```python
def ejemplo(*args, **kwargs):
    print(f"args:", args)
    print(f"kwargs", kwargs)
    
ejemplo("Hola", "Adios", nombre="Paco", apellido="Paquez")
```

```python
# Funciones recursivas

def cuenta_atras(numero):
    print(numero)
    if(numero>1):
        cuenta_atras(numero - 1)
        
cuenta_atras(5)
```

```python
# funciones lambda -> PEQUEÑAS funciones que se escriben normalmente en una sola linea -> SON ANONIMAS

def suma(a,b):
    return a + b

#####################

resultado_suma = lambda a,b: a+b
print(resultado_suma(5,6))
```
