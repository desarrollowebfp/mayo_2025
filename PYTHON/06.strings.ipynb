## Strings


```python
palabra = "Byron"
print(palabra)
```

```python
# Indexación -> Indices
print(palabra[0])
print(palabra[3])

# Indexación negativa
print(palabra[-1])
```

```python
palabra = "Byron"
# rangos de indexacion
# string[start:stop:step]
print(palabra[1:4])
print(palabra[:4])
print(palabra[1:])
print(palabra[::2])
print(palabra[::-1])
```

```python
frase = "Byron es un perro Galgo"
frase_rota = "Que buen día hace"
print(frase.lower())
print(frase.upper())
print(frase.capitalize())
print(frase_rota)
print(frase_rota.strip())  
print(frase.replace("perro", "machango"))
print(frase.find("z"))
# print(frase.index("z"))

# Extra
heroes = ["Batman", "Superman", "Wonderwoman"]
print(" | ".join(heroes))  
```
