```python
heroes = ["Iron Man", "Spider-Man", "Hulk"]
print(heroes[-3])
print(heroes[::-1])

# Función len
print(len(heroes))
```

## Métodos


```python
numeros = [90, 12, 45, 76, 8, 54, 3]
# max y min
print(max(numeros))
print(min(numeros))

# sorted (crea una copia ordenada) y sort (ordena la original)
heroes = ["Iron Man", "Spider-Man", "Hulk", "Black Widow", "Thor", "Hulk"]
heroes_ordenados = sorted(heroes)
print(heroes_ordenados)
heroes.sort()
print(heroes)

# copy (crea una copia independiente)
heroes_copiados = heroes.copy()

# reverse (le damos la vuelta a una lista modificandola)
heroes_copiados.reverse()
print(heroes_copiados)

# count (cuenta cuantas veces aparece un elemento en una lista)
print(heroes.count("Hulk"))
```

```python
# append -> Añadir un elemento al final de la lista
movies = ["Batman"]
movies.append("Sorcerer")
print(movies)

# extend -> Añade varios elementos (lista) al final a una lista
new_movies = ["The Thing", "The Evil Dead", "The Army of Darkness"]
movies.extend(new_movies)
print(movies)

# insert -> Añade un nuevo elemento a un indice concreto
movies.insert(2, "El Viaje de Chihiro")
print(movies)

# pop -> Elimina el ultimo elemento de una lista por defecto -> Si le indicamos un indice cambiamos cual borra
movies.pop(4)
print(movies)

# remove -> Elimina la primera aparición del elemento que le demos por argumento
movies.remove("The Thing")
print(movies)

# clear -> Limpia la lista
movies.clear()
print(movies)
```
