## Funciones Built-in


```python
# print -> mostrar por consola
print("Hola qué tal");
print("Tengo", 2025-1990, "años")
print(f"Tengo {2025-1990} años")
```

```python
# len -> contar elementos/longitudes
print(len("Python"))
print(len([1,2,3]))
```

```python
# type -> consultar el tipo de dato -> <class "float">
print(type("Hola"))
print(type(25))
print(type(50.5))
print(type(True))
print(type(None))
print(type([1,2,3]))
```

```python
# dir -> Mostrar los atributos o métodos que tiene un objeto
texto = "Python"
print(dir(texto))

# help -> muestra la documentacion de una función o método
help(len)
```

```python
# Conversión/Parseo de tipo de datos
print(type(int("123")))
print(type(float("13.5")))
print(type(str(45)))
print(bool(0))
print(bool(1))
print(bool(65))
```

```python
# Funciones numéricas
print(abs(-7))
print(max(3,4,5,6,34,21,65,7,98,9,43,2))
print(min(3,4,5,6,34,21,65,7,98,9,43,2))
print(round(3.14159, 3))
```

```python
# Funciones para listas o secuencias
numeros = [2,4,6,8,10,1,24,9,3]
print(sum(numeros))
print(sorted(numeros))
print(list(reversed(numeros)))
```

```python
# Funciones lógicas
print(any([False, True, False]))
print(all([True, True, True]))
print(all([True, False, True]))
print(all([False, False]))
```

```python
def saludar(nombre, apellidos):
    """
    Esta función te devuelve un saludo por tu nombre pasado por argumento

    Args:
        nombre (str): Este argumento define el nombre para el saludo
        apellidos (str): Este argumento define el apellido para el saludo
    """
    print("Hola", nombre, apellidos)
    
saludar("Antonio", "Byron" )
help(saludar)
```
